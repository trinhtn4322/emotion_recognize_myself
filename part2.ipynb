```python
import numpy as np
import glob
import cv2
import os
```

```python
size = 128
images = []
labels= []

for dic_path in glob.glob('data/*'):
    label=dic_path.split('\\')[-1]
#     print(label)
    for img_path in glob.glob(os.path.join(dic_path,'*.png')):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (size, size))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        images.append(img)
        labels.append(label)

images = np.array(images)
labels = np.array(labels)
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(images,labels, test_size=0.2, random_state=123)
```
